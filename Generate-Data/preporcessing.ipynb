```python
import pandas as pd
```

```python
wajib = pd.read_csv('Matkul Wajib.csv', delimiter=';')
wajib.info()
```

```python
pilihan = pd.read_csv('Matkul Pilihan.csv', delimiter=';')
pilihan.info()
```

```python
rmk = pd.read_csv('RMK Matkul.csv', delimiter=';')
rmk
```

```python
wajib_merge = wajib.merge(rmk, how='left', on ='Course Name')
wajib_merge.info()
```

```python
wajib_merge[wajib_merge['RMK'].isnull()]
```

```python
wajib_merge.head()
```

```python
pilihan_merge = pilihan.merge(rmk, how='left', on ='Course Name')
pilihan_merge.info()
```

```python
pilihan_merge[pilihan_merge['RMK'].isnull()]
```

```python
pilihan_fix = pilihan_merge[['Course Code_x', 'Course Name', 'Credit', 'RMK']].rename(columns={'Course Code_x': 'Course Code'})
pilihan_fix.to_csv('Matkul Pilihan with RMK.csv', index=False, sep=',')
```

```python
pilihan_fix.head()
```

```python
wajib_merge.rename(columns={'Code': 'Course Code', 'Credits': 'Credit'}).to_csv('Matkul Wajib with RMK.csv', index=False, sep=',')
```
